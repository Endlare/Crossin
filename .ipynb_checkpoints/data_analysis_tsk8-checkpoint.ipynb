### 任务1：去除重复行

```python
# -*-coding: utf-8 -*-
import pandas as pd
import numpy as np

jobs_csv=pd.read_csv("houses.csv")
jobs_csv=jobs_csv.drop_duplicates()
jobs_csv
```

### 任务2：数目统计

```python
"""
总房源数量、总小区数量、总版块数量
"""

print("总房源数量: %d" % len(jobs_csv['title'].drop_duplicates()))
print("总小区数量: %d" % len(jobs_csv['xiaoqu_name'].drop_duplicates()))
print("总版块数量: %d" % len(jobs_csv['sub_district_name'].drop_duplicates()))
```

### 任务3：类别统计

```python
"""
不同的户型
不同的装修情况
不同的建造年份
"""

print(jobs_csv.huxing.unique())
print(jobs_csv.zhuangxiu.unique())
print(jobs_csv.buildyear.unique())
```

### 任务4：计算每平米均价

```python
print("上海房产每平米均价：%f" % jobs_csv['up_price'].mean())
```

### 任务5：流行的户型

```python
"""
请统计出不同户型对应的房源数量、平均面积；
并从大到小进行排序，判断出主流户型是什么？平均面积多大？
（任务提示：以 huxing 为分组对象，展示 size 列，统计户型数量和 size 平均数。）
"""

area = pd.DataFrame(jobs_csv[['title','size']].groupby(jobs_csv['huxing']).agg(['count','mean']))
area.sort_values(by=[('size', 'count')],ascending=False)
```

### 任务6：每个小区的房源数量

```python
"""
获取每个小区的房源数量，并按从大到小的顺序排序
"""

title = pd.DataFrame(jobs_csv[['title']].groupby(jobs_csv['xiaoqu_name']).count())
title.sort_values(by=['title'],ascending=False)
```

### 任务7：每个小区均价

```python
price = pd.DataFrame(jobs_csv[['up_price']].groupby(jobs_csv['xiaoqu_name']).mean())
price.sort_values(by=['up_price'],ascending=False)
```

### 任务8：二手房总价前五

```python
jobs_csv.sort_values(by=['price'],ascending=False).head(5)
```

### 统计每十年的房源数量

```python
jobs_csv.buildyear = pd.to_datetime(jobs_csv.buildyear,format="%Y")
title = pd.DataFrame(jobs_csv[['title']].groupby(jobs_csv['buildyear']).count())
title.resample('10A').sum()
```
