### 任务1：去除重复行

```python
import pandas as pd
import numpy as np
import io
import requests

url="https://gitee.com/crossin/data/raw/master/task%207/jobs.csv"
req=requests.get(url).content
jobs_csv=pd.read_csv(io.StringIO(req.decode('utf-8')))
jobs_csv=jobs_csv.drop_duplicates()
jobs_csv
```

### 任务2：查看数据来源

```python
source_val = jobs_csv['source'].value_counts()
source_val = pd.DataFrame(source_val)
source_val
```

### 任务3：工作经验要求

```python
exp_val = jobs_csv['experience'].value_counts()
exp_val = pd.DataFrame(exp_val)
exp_val
```

### 任务4：教育背景要求

```python
edu_val = jobs_csv['education'].value_counts()
edu_val = pd.DataFrame(edu_val)
edu_val
```

### 任务5：不同类型公司的职务数

```python
com_val = jobs_csv['company_type'].value_counts()
com_val = pd.DataFrame(com_val)
com_val
```

### 任务6：计算平均工资

```python
jobs_csv['salary2'].mean()
```

### 任务7：薪资前十位

```python
salary = pd.DataFrame(jobs_csv['salary2'].groupby(jobs_csv['company']).mean())
salary.sort_values(by=['salary2'],ascending=False).head(10)
```

### 任务8：发布职务数目前十位

```python
count = pd.DataFrame(jobs_csv['title'].groupby(jobs_csv['company']).count())
count.sort_values(by=['title'],ascending=False).head(10)
```

### 任务9：薪资前十位的公司类型

```python
salary = pd.DataFrame(jobs_csv['salary2'].groupby(jobs_csv['company_type']).mean())
salary.sort_values(by=['salary2'],ascending=False).head(10)
```
