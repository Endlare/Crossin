### 任务1：创建数据表

```python
import numpy as np
import pandas as pd

d = {"id":[1001,1001,1002,1003,1001,1002,1002,1003,1001,1002,1001,1002],
     "date":['2019-01-01','2019-01-01','2019-01-01','2019-01-01','2019-01-02','2019-01-02',
            '2019-01-02','2019-01-02','2019-01-03','2019-01-03','2019-01-04','2019-01-04'],
     "goods":['A','B','A','B','A','A','B','A','A','B','A','A'],
     "sales":[3000,2000,3000,2000,3000,3000,2000,3000,3000,2000,3000,3000]}
DataFrame = pd.DataFrame(d,index=list(range(12)))
DataFrame
```

### 任务2：统计销售A、B之和

```python
pd.pivot_table(DataFrame,index=["id"],values=["sales"],columns='goods',aggfunc=np.sum)
```

### 任务3：统计销售A+B之和

```python
pd.pivot_table(DataFrame,index=["id"],values=["sales"],columns='goods',aggfunc=np.sum,margins=True)
```

### 任务4：创建数据

```python
d = {"sex":['man','man','women','women','man','women','man','women','women'],
     "age":[15,23,25,17,35,57,24,31,22],
     "smoke":[True, False, False, True, True, False, False, True, False],
     "height":[168,179,181,166,173,178,180,190,160]}
df = pd.DataFrame(d,index=list(range(9)))
df
```

### 任务5：查看男女平均身高

```python
pd.pivot_table(df, index=['sex'],values=['age','height'])
```

### 任务6：统计各个性别抽烟人数

```python
pd.crosstab(df.sex,df.smoke)
```

### 任务7：统计各个年龄段抽烟人数

```python
pd.crosstab(df.age,df.smoke)
```
